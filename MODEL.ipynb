```python
import pandas as pd
```

```python
movie_df= pd.read_csv('movie_metadata.csv')
movie_df.head()
```

```python
summary= movie_df.describe(include= 'all')
summary
```

```python
missing_vals= movie_df.isnull().sum()
missing_vals
```

```python
#Impute missing values for numeric features with the mean
numeric_features= ['num_critic_for_reviews', 'duration', 'director_facebook_likes', 
                    'actor_3_facebook_likes', 'actor_1_facebook_likes', 'gross', 
                    'num_user_for_reviews', 'budget', 'actor_2_facebook_likes', 'aspect_ratio']
for feature in numeric_features:
    movie_df[feature].fillna(movie_df[feature].mean(), inplace=True)

#Remove rows with missing values in critical features
critical_features= ['genres', 'movie_title', 'director_name', 'actor_1_name', 'actor_2_name', 'actor_3_name']
movie_df.dropna(subset=critical_features, inplace=True)

#Use default values for categorical features
categorical_features= ['color', 'language', 'country', 'content_rating']
for feature in categorical_features:
    movie_df[feature].fillna('Unknown', inplace=True)

#Verify that all missing values are handled
missing_values_after = movie_df.isnull().sum()


missing_values_after
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

#Combine selected features into a single string for each movie
movie_df['combined_features'] = movie_df['genres'] + ' ' + \
                                  movie_df['director_name'] + ' ' + \
                                  movie_df['actor_1_name'] + ' ' + \
                                  movie_df['actor_2_name'] + ' ' + \
                                  movie_df['actor_3_name']

#Initialize the TF-IDF Vectorizer
tfidf_vectorizer= TfidfVectorizer(stop_words='english')

#Apply the vectorizer on the combined features
tfidf_matrix= tfidf_vectorizer.fit_transform(movie_df['combined_features'])

#Display the shape of the resulting matrix
tfidf_matrix_shape = tfidf_matrix.shape

tfidf_matrix_shape
```
